```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from pathlib import Path
```

```python
import os
print("Working directory:", os.getcwd())
```

```python
# Example simulation
data_path = Path("../data/netsim/sims/") #change to make relative to the root directory, config.py
example_file = data_path / "sim1.mat"
mat = loadmat(example_file)

# Extract values
ts = mat["ts"]
net = mat["net"]
Ntimepoints = mat["Ntimepoints"][0][0]
Nnodes = mat["Nnodes"][0][0]

# Get data for subject 1
ts1 = ts[:Ntimepoints, :]
net1 = net[0, :, :]

# Plot time series for first 5 nodes
plt.figure(figsize=(12, 6))
for i in range(5):
    plt.plot(ts1[:, i], label=f"Node {i+1}")
plt.title("Node Time Series (First Subject, First 5 Nodes)")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ground-truth network for subject 1
plt.figure(figsize=(6, 5))
plt.imshow(net1, cmap="viridis", interpolation="nearest")
plt.title("Ground-Truth Connectivity Matrix (Subject 1)")
plt.colorbar(label="Connection Strength")
plt.xlabel("Node")
plt.ylabel("Node")
plt.tight_layout()
plt.show()
```
